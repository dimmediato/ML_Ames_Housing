```python
import pandas as pd
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor, GradientBoostingClassifier
pd.set_option('display.max_columns', 500)
housing_factor = pd.read_pickle("housing_factor.pkl")
housing_factor
```

```python
label_encoder = LabelEncoder()
years = housing_factor[["YrSold",'GarageYrBlt','MoSold','YearBuilt', 'YearRemodAdd']].copy()
housing_factor_dummified1 = housing_factor.drop(['PID', "YrSold",'GarageYrBlt','MoSold','YearBuilt', 'YearRemodAdd'], axis=1)

housing_factor_dummified1= housing_factor_dummified1.apply(label_encoder.fit_transform)
housing_factor_dummified1
housing_factor_dummified1
```

```python
housing_numeric = pd.read_pickle("housing_numeric.pkl")
housing_numeric
```

```python
housing_factor_dummified1 = pd.concat([housing_numeric, housing_factor_dummified1], axis=1)
housing_factor_dummified1 = pd.concat([housing_factor_dummified1, years], axis=1)
```

```python
housing_factor_dummified1
```

```python
gbm = GradientBoostingClassifier()
housing_factor_dummified2 = housing_factor_dummified1.drop(['SalePrice'], axis=1)
housing_factor_dummified2['GarageYrBlt'] = housing_factor_dummified2['GarageYrBlt'].replace('NoGarage', 1970)
```
