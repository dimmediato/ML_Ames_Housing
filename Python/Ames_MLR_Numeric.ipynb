```python
import pandas as pd
import numpy as np
from scipy.stats import norm
pd.set_option('display.max_columns', 500)
housing = pd.read_csv("~/Documents/NYCDSA/Third Project/ML_Ames_Housing/data/Ames_Housing_Price_Data.csv", index_col=0)

housing_numeric = pd.read_pickle('housing_numeric.pkl')
housing_numeric = housing_numeric.drop(['PID'], axis=1)
housing_numeric
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(housing_numeric, housing['SalePrice'])
SalePrice = housing['SalePrice']

import matplotlib.pyplot as plt
residuals = SalePrice - lm.predict(housing_numeric)
plt.hist(residuals)
```

```python
import matplotlib.pyplot as plt
residuals = SalePrice - lm.predict(housing_numeric)
plt.hist(residuals)
```

```python
print('R^2 is equal to %.3f' %(lm.score(housing_numeric,SalePrice))) #By default, score gives R^2. 
print('RSS is equal to %.3f' %(np.sum(residuals**2)))
print('The intercept is %.3f' %(lm.intercept_))
print('The slopes are %s' %(lm.coef_))
```

```python
print('negative slope column names %s' %(housing_numeric.columns[lm.coef_ < 0]))
print('positive slope column names %s' %(housing_numeric.columns[lm.coef_ > 0]))
```

```python
log_SalePrice = np.log(SalePrice) #Take log of DV
lm.fit(housing_numeric, log_SalePrice) #LM of log of DV
residuals = log_SalePrice - lm.predict(housing_numeric)
plt.hist(residuals)
```

```python
print('R^2 is equal to %.3f' %(lm.score(housing_numeric,log_SalePrice))) #By default, score gives R^2. 
print('RSS is equal to %.3f' %(np.sum(residuals**2)))
print('The intercept is %.3f' %(lm.intercept_))
print('The slopes are %s' %(lm.coef_))
```
